```python
df=pd.read_csv(r'C:\Users\HP\Desktop\Pandas\Data.csv')
df
```

```python
sns.lineplot(x='Roll',y='Height',data=df,marker=".",ms=22,mec="red",mfc="w")
plt.grid()
plt.legend("Height",loc="lower right")
plt.show()
```

```python
#Themes: There are five themes available in Seaborn.(darkgrid,whitegrid,dark,white,ticks)
plt.subplot(2,2,1)
sns.lineplot(x='Roll',y='Height',data=df,marker=".",ms=22,mec="blue",mfc="w",color="green")
plt.legend("Height",loc="lower right")
plt.title("Ticks")
sns.set_style('ticks')


plt.subplot(2,2,2)
sns.lineplot(x='Roll',y='Height',data=df,marker="o",ms=12,mec="red",mfc="w",color="blue")
plt.legend("Height",loc="lower right")
plt.title("dark")
sns.set_style('dark')

plt.subplot(2,2,3)
sns.lineplot(x='Roll',y='Height',data=df,marker=".",ms=22,mec="hotpink",mfc="w")
plt.legend("Height",loc="lower right")
plt.title("whitegrid")
sns.set_style('whitegrid')

plt.subplot(2,2,4)
sns.lineplot(x='Roll',y='Height',data=df,marker="*",ms=12,mec="green",mfc="w")
plt.legend("Height",loc="lower right")
plt.title("darkgrid")
sns.set_style('darkgrid')
plt.suptitle("Themes of Seaborn")
plt.show()
```

```python
# Removing spines
sns.lineplot(x='Roll',y='Height',data=df,marker=".",ms=22,mec="red",mfc="w")
plt.legend("Height",loc="lower right")
plt.suptitle("Removing of Spines")
sns.set_style("white")
sns.despine()
```

```python
# figure size
plt.figure(figsize = (2, 2))
sns.lineplot(x='Roll',y='Height',data=df,marker=".",ms=22,mec="red",mfc="w")
sns.despine()
plt.show()
```

```python
# Scaling the plots
plt.figure(figsize = (4, 4))
sns.lineplot(x='Roll',y='Height',data=df,marker=".",ms=22,mec="red",mfc="w")
sns.set_context("paper",font_scale=1, rc=None)
plt.show()
```

# subplots

```python
# add_axes() 
def graph():
    sns.lineplot(x='Roll',y='Height',data=df)
fig = plt.figure(figsize =(5, 4)) 
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
graph()
ax2 = fig.add_axes([0.5, 0.4, 0.3, 0.3])
graph()

plt.show()
```

```python
# subplot() 

def graph():
    sns.lineplot(x='Roll',y='Height',data=df) 
plt.subplot(121)
graph()
plt.subplot(122)
graph()
sns.despine()
plt.show()
```

```python
#  subplot2grid() 
def graph():
    sns.lineplot(x='Roll',y='Height',data=df) 
axes1 = plt.subplot2grid (
  (7, 3), (0, 0), rowspan = 2,  colspan = 1) 
graph()
  
axes2 = plt.subplot2grid (
  (7, 3), (2, 0), rowspan = 2, colspan = 1) 
graph()
    
axes3 = plt.subplot2grid (
  (7, 3), (4, 0), rowspan = 2, colspan = 1)
graph()
plt.show()
```
